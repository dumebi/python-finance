```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%load_ext autoreload
%autoreload 2

### load in and normalize the dataset
finance_data = pd.read_csv('AUDJPY.csv')
finance_data = finance_data.drop(['symbol','tradingDay'], axis = 1)
finance_data
```

```python
finance_data['timestamp'] = pd.to_datetime(finance_data['timestamp'])
# plt.plot()

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111)
ax.plot(finance_data['timestamp'], finance_data['close'])


plt.xlabel('time period')
plt.ylabel('normalized series value')
plt.show()
```

```python
features = finance_data.drop(['high','low','close', 'volume'], axis = 1)
features = features.set_index('timestamp')
close = finance_data.drop(['high','low','open', 'volume', 'timestamp'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, close, test_size=0.4, shuffle=True, random_state=101)
```
